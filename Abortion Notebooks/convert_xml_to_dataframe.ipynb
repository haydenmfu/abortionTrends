```python
# Importing our required libraries
import os
import pandas as pd
from lxml import etree
from bs4 import BeautifulSoup
import random
```

```python
# Replace the path below with the dataset which you would like to use as input for the script
dataset_directory = 'C:\\Users\\hayden\\Downloads\\XML_DATA exported_(10)'
input_files = os.listdir(dataset_directory)
```

```python
with open(dataset_directory + "\\1023550511.xml", 'r') as f:
    data = f.read()
#print(data) 

Bs_data = BeautifulSoup(data, "xml")

#print(Bs_data)
 
# Finding all instances of tag
# `unique`
b_unique = Bs_data.find_all('LastName')
 
print(b_unique)

b_unique = Bs_data.find_all('FirstName')
 
print(b_unique)
 
# Using find() to extract attributes
# of the first instance of the tag
#b_name = Bs_data.find('child', {'name':'Frank'})
 
#print(b_name)
 
# Extracting the data stored in a
# specific attribute of the
# `child` tag
#value = b_name.get('test')
 
#print(value)
```

```python
firstnames = [""]*len(input_files)
lastnames = [""]*len(input_files)
dates = [""]*len(input_files)
titles = [""]*len(input_files)
subtitles = [""]*len(input_files)
language = [""]*len(input_files)
abstracts = [""]*len(input_files)

        
for i in range(len(input_files)):
    with open(dataset_directory+"\\" + input_files[i], 'r') as f:
        data = f.read()
        Bs_data = BeautifulSoup(data, "xml")
        if len(Bs_data.find_all('FirstName')) == 0:
            firstnames[i] = "None"
        else:
            firstnames[i] = Bs_data.find_all('FirstName')
            firstnames[i] = firstnames[i][0].get_text()
        if len(Bs_data.find_all('LastName')) == 0:
            lastnames[i] = "None"
        else:
            lastnames[i] = Bs_data.find_all('LastName')
            lastnames[i] = lastnames[i][0].get_text()
        if len(Bs_data.find_all('NumericDate')) == 0:
            dates[i] = "None"
        else:
            dates[i] = Bs_data.find_all('NumericDate')
            dates[i] = dates[i][0].get_text()
        if len(Bs_data.find_all('TitleAtt')) == 0:
            titles[i] = "None"
        else:
            titles[i] = Bs_data.find_all('TitleAtt')
            titles[i] = titles[i][0].get_text()
            titles[i] = titles[i].strip()
        if len(Bs_data.find_all('SubTitleAtt')) == 0:
            subtitles[i] = "None"
        else:
            subtitles[i] = Bs_data.find_all('SubTitleAtt')
            subtitles[i] = subtitles[i][0].get_text()
            subtitles[i] = subtitles[i].strip()
        if len(Bs_data.find_all('RawLang')) == 0:
            language[i] = "None"
        else:
            language[i] = Bs_data.find_all('RawLang')
            language[i] = language[i][0].get_text()
            language[i] = language[i].strip()
        if len(Bs_data.find_all('Abstract')) == 0:
            abstracts[i] = "None"
        else:
            abstracts[i] = Bs_data.find_all('Abstract')
            abstracts[i] = abstracts[i][0].get_text()
            abstracts[i] = abstracts[i].strip()
        
        
```
